```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

os.getcwd()
```

```python
##loading the dataset "golf" provided from kaggle.com
Driving_Distance = pd.read_excel('Golf.xlsx', sheet_name='Golf')
print(Driving_Distance)
```

```python
df = pd.DataFrame(Driving_Distance)

##Average drive distance without coating 270.275
print("mean: ",df["Without Coating"].mean())

##Average drive distance with coating 267.5
print("mean: ",df["With Coating"].mean())
```

```python
##Create Bar Chart

fig = plt.figure(figsize=(8,6))

Ball_Type = ["Without Coating", "With Coating"]
Average_Drive = [270.275, 267.5]
Positions = [0,1]

plt.title("Average Drive Distance (yards)")
plt.bar(Positions, Average_Drive, width = .5)

plt.xticks(Positions, Ball_Type)

plt.show()
```

```python
##Create Line Chart

plt.plot(df)
plt.title("Average Drive Distance (yards)")
plt.ylabel("Drive Distance")
plt.xlabel("Orange is with coating blue is without coating")
plt.ylim(240,290)
```

```python
##Create Pie Chart

Average_Drive_Distance = np.array([270.275, 267.5])
Ball_Type = ["Without Coating", "With Coating"]

plt.pie(Average_Drive_Distance, labels = Ball_Type)
plt.show()
```
